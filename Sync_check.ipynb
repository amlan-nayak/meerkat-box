```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from pathlib import Path
```

```python
k = 'ZU_2021_1'
ModelData = '/media/amlan/Data/Thesis Data/Processed Data/'+k+'/ModelData/'



Modelpaths = os.listdir(ModelData)
```

```python
import datetime as dt
#%matplotlib widget
figs = []
for i in Modelpaths:
    file = i
    #print(f'Dataframe {i} reading initiated')
    Data = pd.read_csv(ModelData+file,index_col=0)
    Data['Timestamp'] = pd.to_datetime(Data['Timestamp'])
    Data1 = Data.copy()
    #Data = Data[(Data['Timestamp']>dt.datetime(2021, 5, 16, 9,47,0)) & (Data['Timestamp']<dt.datetime(2021, 5, 16, 9, 51,30))]
   # print(f'Read dataframe\n', Data.head())
    fig,ax = plt.subplots(figsize=(16,9))
    x = Data['Timestamp']
    y = Data['X_Mean']
    z = Data['VeDBA']
    ax.plot(x, y,label='X Mean')
    ax.plot(x, z, label='VeDBA')
    ax.set(xlabel='Time', ylabel='VeDBA')
    Data_audits = Data.groupby('Behavior')
    for state, values in Data_audits:
        ax.scatter(values['Timestamp'], [1.4]*len(values['Timestamp']), label=state)
    ax.legend()
    
    plt.show()
```
